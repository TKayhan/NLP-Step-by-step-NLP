```python
import pandas as pd
import warnings

warnings.simplefilter("ignore", UserWarning)
warnings.simplefilter("ignore", FutureWarning)
warnings.simplefilter("ignore", DeprecationWarning)
```

```python
df=pd.read_csv("IMDB Dataset.csv")
```

```python
df.head()
```

```python
df["review"]= df["review"].str.lower() #kucuk harfe donusturduk
df["review"]= df["review"].str.replace('[^\w\s]','')#noktalama işaretlerini kaldır
df["review"]= df["review"].str.replace('\n','') # slash n leri bul kaldır boşlukları bul kaldır
df["review"]= df["review"].str.replace('\d+','') #rakamları bul kaldır
df["review"]= df["review"].str.replace('\r','') #satırbaslarını kaldır
```

```python
df.head()
```

```python
mapping = {'positive': 1, 'notr': 0, 'negative': -1}

# 'duygu' sütununu dönüştürelim
df['sentiment'] = df['sentiment'].map(mapping)

# Sonuçları yazdıralım
print(df)
```

```python
import numpy as np
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split
```

```python
df = df[["review", "sentiment"]]
x = np.array(df["review"])
y = np.array(df["sentiment"])

cv = CountVectorizer()
x = cv.fit_transform(x)
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                test_size=0.2, 
                                                random_state=42)
```
